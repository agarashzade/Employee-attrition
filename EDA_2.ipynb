### Cleaning the data

```python
#To check the correlations, we add a dummy variable for Attrition
df['d_Attrition']=df['Attrition'].map({'Yes': 1, 'No': 0})
```

```python
#The column is always 1 and does not serve for any purpose, we drop it
df=df.drop('EmployeeCount',axis=1)
```

```python
df['BusinessTravel'].value_counts()
```

```python
#We reword the Business Travel column
df['BusinessTravel']=df['BusinessTravel'].map({'Travel_Rarely': 'Rare','Travel_Frequently': 'Frequent','Non-Travel':'Never'})
df=df.rename(columns={'BusinessTravel': 'TravelFrequency'})
```
