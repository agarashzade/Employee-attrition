### Cleaning the data and feature engineering

```python
#To check the correlations, I add a dummy variable for Attrition
df['d_Attrition']=df['Attrition'].map({'Yes': 1, 'No': 0})
```

```python
#The column is always 1 and does not serve for any purpose, we drop it
df=df.drop('EmployeeCount',axis=1)
```

```python
#We reword the Business Travel column
df['BusinessTravel']=df['BusinessTravel'].map({'Travel_Rarely': 'Rare','Travel_Frequently': 'Frequent','Non-Travel':'Never'})
df=df.rename(columns={'BusinessTravel': 'TravelFrequency'})
```

### We join the dictionaries for some columns: 

**Education:**
1 'Below College' 2 'College' 3 'Bachelor' 4 'Master' 5 'Doctor'

**EnvironmentSatisfaction:**
1 'Low' 2 'Medium' 3 'High' 4 'Very High'

**JobInvolvement:**
1 'Low' 2 'Medium' 3 'High' 4 'Very High'

**JobSatisfaction:**
1 'Low' 2 'Medium' 3 'High' 4 'Very High'

**PerformanceRating:**
1 'Low' 2 'Good' 3 'Excellent' 4 'Outstanding'

**RelationshipSatisfaction:**
1 'Low' 2 'Medium' 3 'High' 4 'Very High'

**WorkLifeBalance:**
1 'Bad' 2 'Good' 3 'Better' 4 'Best'

```python
dict_education=pd.DataFrame({'Education':[1,2,3,4,5],'Education_desc':['Below College','College','Bachelor','Master','Doctor']})
dict_env_sat=pd.DataFrame({'EnvironmentSatisfaction':[1,2,3,4],'Env_sat_desc':['Low','Medium','High','Very High']})
dict_job_involve=pd.DataFrame({'JobInvolvement':[1,2,3,4],'Job_involve_desc':['Low','Medium','High','Very High']})
dict_job_sat=pd.DataFrame({'JobSatisfaction':[1,2,3,4],'Job_sat_desc':['Low','Medium','High','Very High']})
dict_perf_rating=pd.DataFrame({'PerformanceRating':[1,2,3,4],'Perf_rating_desc':['Low','Good','Excellent','Outstanding']})
dict_relation_sat=pd.DataFrame({'RelationshipSatisfaction':[1,2,3,4],'Relation_sat_desc':['Low','Medium','High','Very High']})
dict_work_life=pd.DataFrame({'WorkLifeBalance':[1,2,3,4],'Work_life_desc':['Bad','Good','Better','Best']})
```

```python
df=df.merge(dict_education,how='left',left_on='Education',right_on='Education')
df=df.merge(dict_env_sat,how='left',left_on='EnvironmentSatisfaction',right_on='EnvironmentSatisfaction')
df=df.merge(dict_job_involve,how='left',left_on='JobInvolvement',right_on='JobInvolvement')
df=df.merge(dict_job_sat,how='left',left_on='JobSatisfaction',right_on='JobSatisfaction')
df=df.merge(dict_perf_rating,how='left',left_on='PerformanceRating',right_on='PerformanceRating')
df=df.merge(dict_relation_sat,how='left',left_on='RelationshipSatisfaction',right_on='RelationshipSatisfaction')
df=df.merge(dict_work_life,how='left',left_on='WorkLifeBalance',right_on='WorkLifeBalance')
```
